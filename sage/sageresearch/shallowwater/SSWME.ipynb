```python
import sageresearch.shallowwater.spherical_shallow_water_moment_equations as sswme
import sageresearch.utils.symbolic_vector_matrix as svm
```

# Spherical Shallow Water Moment Equations

```python
num_moments = 0
tuple_ = sswme.get_shallow_water_moment_equations(num_moments)
f_theta_p = tuple_[0]
f_phi_p = tuple_[1]
G_theta_p = tuple_[2]
G_phi_p = tuple_[3]
s_p = tuple_[4]
A_theta_p = tuple_[5]
A_phi_p = tuple_[6]
print("f_theta")
print(f_theta_p)
print("f_phi")
print(f_phi_p)
print("G_theta")
print(G_theta_p)
print("G_phi")
print(G_phi_p)
print("source")
print(s_p)
```

```python
n = svm.get_vector_variable("n", 3)
eig = (n[1] * A_theta_p + n[2] * A_phi_p).eigenvalues()
print(eig)
```
